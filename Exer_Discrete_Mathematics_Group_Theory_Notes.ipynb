Definición 1. Un grupo es un conjunto en el cual existe un elemento especial o neutro $e$ con una operación binaria asociativa, tal que:
> (i) $e*a=a$ $∀$ $G$
>
> (ii) $∀ a ∈ G $ $∃$ $b ∈ G $ $|$ $b*a=e$

# 1.1.1-1. For each fixed integer $n > 0$, prove that $Zn$, the set of integers modulo n is a group under $+$, where one defines $[a] + [b] = [a + b]$. (The elements of $Zn$ are the congruence classes $[a], a ∈ Z$. The congruence class $[a]$ is

        {x ∈ Z : x ≡ a (mod n)} = {a + kn : k ∈ Z}.

> **Be sure to show that this addition is well defined. Conclude that for every integer n > 0 there is a group with n elements.**
---
Teniendo en cuenta que la operación binaria está bien definida, puesto que cada elemento del grupo debe pertencer tanto a los operandores como a los resultados, definiendo la operación como $[a] + [b] = [a + b]$.

Tomando 2 elementos genéricos de las clases residuales de la forma $a',a ∈ [a]$ y $b', b ∈ [b]$ se tendrá que: $[a + b] = [a' + b']$ teniendo en cuenta las clases de congruencia del modo:

    [a + b] -> {x ∈ Z : x ≡ (a + b) mod n}

Propiedades:

**Asociatividad:** 

> $[a + b] + [c] = [a] + [b + c]$
>
> $... ... ... ... = [a + b +c]$
>
> $... ... ... ... = [[a + b] + c]$
>
> $[a + b] + [c] = [a + b] + [c]$

**Inversa:**

> $[a]^{-1} = -[a] ∈ G$
>
> $[a] + [a]^{-1} = [a]-[a]$
>
> $[a] + [a]^{-1} = [0]$
>
> $[0] = [0]$

**Elemento neutro:**

> $[0] ∈ G$
>
> $[0] + [a] = [a]$ Por esto:
>
> $e = [0]$

Con esto, por la definición del algoritmo de división se tiene que $x - (a + b) = nk$ el cual permite aplicar la Definición de Clase Residual de la siguiente forma (tomando enteros cuales quiera $w$, $q$):

> $a = [a] mod (n)$
>
> $a - [a] = nw$
>
> $a = nw + [a]$

> $b = [b] mod (n)$
>
> $b - [b] = nq$
>
> $b = nq + [b]$

Ahora, reemplazando en el algoritmo mencionado anteriormente, se tendrá:

> $x - (([a] + nw) + ([b] + nq)) = nk$
> 
> $x = nk + [a] + nw + [b] + nq$
>
> $x = ([a] + [b]) + n(k +w +q)$

Lo cual se interpretará como $x ≡ ([a] + [b]) mod (n)$, por cual, se podrá afirmar que $x ∈ ([a] + [b])$, confirmando que $+$ es una Operación Bien Definida.
